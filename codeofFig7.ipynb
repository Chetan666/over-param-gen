```python
from google.colab import files
# Compute A^T A
AtA = A.T @ A  # (p x n) @ (n x p) = (p x p)

# Plot the result
plt.figure(figsize=(6, 5))
plt.imshow(AtA, cmap='viridis', interpolation='nearest')
plt.xlabel('j')
plt.ylabel('i')
plt.title(r'Result of $\delta^T \delta$ (size {}x{})'.format(p, p))
plt.colorbar(label=r'$(\delta^T \delta)_{ij}$')
# Save the plot to a PDF file
plt.savefig("ATA.pdf")
files.download("ATA.pdf")
plt.show()





```

```python
# Compute A A^T
AAt = A @ A.T  # (n x p) @ (p x n) = (n x n)

# Plot the result
plt.figure(figsize=(6, 5))
plt.imshow(AAt, cmap='plasma', interpolation='nearest')
plt.xlabel('j ')
plt.ylabel('i ')
plt.title(r'Result of $\delta \delta^T$ (size {}x{})'.format(n, n))
plt.colorbar(label=r'$(\delta \delta^T)_{ij}$')
# Save the plot to a PDF file
plt.savefig("AAT.pdf")
files.download("AAT.pdf")
plt.show()
```
